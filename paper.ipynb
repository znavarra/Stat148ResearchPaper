```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('final.csv')
data
```

```python
def plot_grouped_bar_chart(df, group_col, value_col, title='Average Bar Chart'):
    # Group by the specified column and calculate the mean
    grouped = df.groupby(group_col)[value_col].mean().sort_values()

    # Plot
    plt.figure(figsize=(10, 6))
    grouped.plot(kind='bar', color='skyblue', edgecolor='black')

    plt.title(title)
    plt.xlabel(group_col)
    plt.ylabel(f'Average {value_col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
data.groupby('treatment')['avg'].mean()
```
